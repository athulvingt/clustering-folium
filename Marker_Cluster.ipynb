# **Marker Cluster**
This notebook Clusters 50000 markers on a folium-leaflet map

## Download dataset
The dataset used for this purpose is collected from Weeplaces, a website that aims to visualize users' check-in activities in location-based social networks (LBSN). All the crawled data is originally generated in Foursquare. This dataset contains 7,658,368 check-ins generated by 15,799 users over 971,309 locations. **Note that this dataset is released solely for research purpose.**

```
# import
import pandas as pd
import folium
from folium import plugins 
```

```
df = pd.read_csv('weeplaces/weeplace_checkins.csv', encoding='utf-8')
data = df.sample(50000)
data.describe()
```

```
center = [data.lat.mean(), data.lon.mean()]
```

The Tiles I have used here is Esri.WorldGray.Canvas, If you want to try different tiles check this link [leaflet-providers](https://leaflet-extras.github.io/leaflet-providers/preview/).

```
# populating map

map1 = folium.Map(location= center, zoom_start=4,
                  tiles='https://server.arcgisonline.com/ArcGIS/rest/services/Canvas/World_Light_Gray_Base/MapServer/tile/{z}/{y}/{x}',
                  attr= 'Tiles &copy; Esri &mdash; Esri, DeLorme, NAVTEQ',)

incidents = plugins.MarkerCluster(chunkedLoading = True).add_to(map1)


for lat, lng in zip(data.lat,data.lon):
  folium.Marker(location = [lat, lng],icon=folium.Icon(color='black', icon='ok-sign')).add_to(incidents)
 
map1
```

```
# To save map
map1.save("map50000.html")
```
